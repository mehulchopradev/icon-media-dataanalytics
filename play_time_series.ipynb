```python
import pandas as pd
import matplotlib.pyplot as plt
```

# Trend time series data

```python
path = 'data/trend-data.csv'
d1 = pd.read_csv(path, parse_dates=['date'], index_col='date')
d1
```

# Seasonal time series data

```python
path = 'data/seasonal-data.csv'
d2 = pd.read_csv(path, parse_dates=['date'], index_col='date')
d2
```

# Trend + Seasonal time series data

```python
path = 'data/trend-seasonal-data.csv'
d3 = pd.read_csv(path, parse_dates=['date'], index_col='date')
d3
```

```python
figure, axes = plt.subplots(1, 3, figsize=(18, 4))
d1.plot(ax=axes[0], title='Trend data', legend=False)
d2.plot(ax=axes[1], title='Seasonal data', legend=False)
d3.plot(ax=axes[2], title='Trend + Seasonal data', legend=False)
```
