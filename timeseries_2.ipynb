```python
import pandas as pd
from pandas import DataFrame
```

```python
path = 'data/sales_data_datetime.csv'
d1 = pd.read_csv(path, parse_dates=['date'], index_col='date')
d1
```

```python
# Resampling of time series data
# Downsampling (decreasing the number of samples in time series data by increasing the frequency)
d1.resample('2H', base=1).sum()
```

```python
d1.resample('2H', base=1).agg(['sum', 'mean'])
```

```python
# upsampling (increase the number of samples in time series data by decreasing the frequency)
d2 = DataFrame({
    'year': ['2019-01-01', '2020-01-01', '2021-01-01'],
    'sales': [56000, 60000, 570000]
})
d2.index = pd.to_datetime(d2['year'])
del d2['year']
d2
```

```python
d2.resample('Q').ffill()
```

```python
sales_data = pd.read_csv('data/sales_data_date.csv', parse_dates=['date'], index_col='date')
sales_data
```

```python
price_data = pd.read_csv('data/price_data_date.csv', parse_dates=['date'], index_col='date')
price_data
```

```python
# upsampling
d5 = price_data.resample('M').ffill()
d5
```

```python
d6 = pd.concat([sales_data, d5], axis=1)
d6['total'] = d6['num_sold'] * d6['price']
d6
```
