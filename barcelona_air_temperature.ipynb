```python
import pandas as pd
```

```python
path = 'data/temperaturesbarcelonadesde1780.csv'
d1 = pd.read_csv(path)
d1
```

```python
d1.set_index('Any', inplace=True)
d1.index.name = 'Year'
```

```python
d1
```

```python
d1['yearly_avg'] = d1.mean(axis=1)
d1
```

```python
d2 = d1[['yearly_avg']]
d2
```

```python
d2.plot()
```

```python
# rolling averages ( rolling mean, rolling window)
```

```python
# Simple moving average
```

```python
d2['rolling_3'] = d2.yearly_avg.rolling(3, min_periods=1).mean()
d2
```

```python
d2['rolling_10'] = d2.yearly_avg.rolling(10, min_periods=1).mean()
d2
```

```python
d2['rolling_20'] = d2.yearly_avg.rolling(20, min_periods=1).mean()
d2
```

```python
d2[['rolling_20']].plot()
```
